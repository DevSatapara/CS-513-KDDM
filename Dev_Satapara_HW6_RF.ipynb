```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
header_names = ["SampleCode", "ClumpThickness", "UniformityCellSize", "UniformityCellShape", "MarginalAdhesion", "SingleEpithelialCellSize", "BareNuclei", "BlandChromatin", "NormalNucleoli", "Mitoses", "Class"]
```

```python
df =  pd.read_csv('breast-cancer-wisconsin.csv',header=None,skiprows=1, names=header_names,na_values='?')
df
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df['Class'] = df['Class'].map({2: 'Benign', 4: 'Malignant'}).fillna(df['Class'])
df
```

```python
X = df.drop(columns=['SampleCode', 'Class'])
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
RF = RandomForestClassifier()
RF = RF.fit(X_train,y_train)

y_pred = RF.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)*100
print(accuracy)
```

```python
print(pd.Series(RF.feature_importances_,index=X.columns))
```
