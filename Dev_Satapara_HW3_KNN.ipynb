```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv('C:\\Users\\satap\\OneDrive\\Desktop\\STEVENS\\CS-513\\breast-cancer-wisconsin.csv')
```

```python
header_names = ["SampleCode", "ClumpThickness", "UniformityCellSize", "UniformityCellShape", "MarginalAdhesion", "SingleEpithelialCellSize", "BareNuclei", "BlandChromatin", "NormalNucleoli", "Mitoses", "Class"]
```

```python
df =  pd.read_csv('C:\\Users\\satap\\OneDrive\\Desktop\\STEVENS\\CS-513\\breast-cancer-wisconsin.csv',header=None,skiprows=1, names=header_names,na_values='?')
df
```

```python
df.dropna(axis=0,inplace=True)
```

```python
df
```

```python
df['Class'] = df['Class'].map({2: 'Benign', 4: 'Malignant'}).fillna(df['Class'])
```

```python
df
```

```python
X = df.drop(columns=['SampleCode', 'Class'])
y = df['Class']
```

```python
X = pd.DataFrame(MinMaxScaler().fit_transform(X), columns=X.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
knn_3 = KNeighborsClassifier(n_neighbors=3)
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_10 = KNeighborsClassifier(n_neighbors=10)
```

```python
knn_3.fit(X_train, y_train)
```

```python
knn_5.fit(X_train,y_train)
```

```python
knn_10.fit(X_train,y_train)
```

```python
y_pred_3 = knn_3.predict(X_test)
y_pred_5 = knn_5.predict(X_test)
y_pred_10 = knn_10.predict(X_test)
```

```python
conf_matrix_3 = confusion_matrix(y_test, y_pred_3)
print(conf_matrix_3)
```

```python
conf_matrix_5 = confusion_matrix(y_test, y_pred_5)
print(conf_matrix_5)
```

```python
conf_matrix_10 = confusion_matrix(y_test, y_pred_10)
print(conf_matrix_10)
```

```python
accuracy_3 = accuracy_score(y_test, y_pred_3) * 100
print(f'Accuracy: {accuracy_3:.2f}%')
```

```python
accuracy_5 = accuracy_score(y_test, y_pred_5) * 100
print(f'Accuracy: {accuracy_5:.2f}%')
```

```python
accuracy_10 = accuracy_score(y_test, y_pred_10) * 100
print(f'Accuracy: {accuracy_10:.2f}%')
```
